```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons, make_blobs
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import NearestNeighbors
import pandas as pd
from tqdm import tqdm
```

```python


class DBSCAN:
    def __init__(self, eps=0.5, min_samples=5):
        self.eps = eps
        self.min_samples = min_samples
        self.labels_ = None

    def fit(self, X):
        n_samples = X.shape[0]
        self.labels_ = np.full(n_samples, -1)  # -1 means unvisited/noise
        cluster_id = 0

        # Precompute neighbors for efficiency
        neighbors_model = NearestNeighbors(radius=self.eps).fit(X)
        neighborhoods = neighbors_model.radius_neighbors(X, return_distance=False)

        for i in range(n_samples):
            if self.labels_[i] != -1:  # Already visited
                continue

            # Find all points reachable from the current point
            neighbors = neighborhoods[i]

            if len(neighbors) < self.min_samples:
                self.labels_[i] = -1  # Mark as noise
                continue

            # Start a new cluster
            cluster_id += 1
            self.labels_[i] = cluster_id

            # Process the neighborhood
            seed_set = set(neighbors)
            seed_set.discard(i)  # Remove the current point

            while seed_set:
                j = seed_set.pop()

                if self.labels_[j] == -1:  # If noise, add to cluster
                    self.labels_[j] = cluster_id

                if self.labels_[j] != -1:  # Already assigned to a cluster
                    continue

                self.labels_[j] = cluster_id

                # Find neighbors of this point
                j_neighbors = neighborhoods[j]

                if len(j_neighbors) >= self.min_samples:
                    # Add new neighbors to the seed set
                    seed_set.update(j_neighbors)

        return self

    def fit_predict(self, X):
        self.fit(X)
        return self.labels_

# Generate sample datasets
def generate_datasets():
    # Dataset 1: Two moons with noise
    X1, y1 = make_moons(n_samples=1000, noise=0.05, random_state=0)
    X1 = StandardScaler().fit_transform(X1)

    # Dataset 2: Blobs with varying density
    X2, y2 = make_blobs(n_samples=300, centers=3, cluster_std=[1.0, 2.5, 0.5], random_state=0)
    X2 = StandardScaler().fit_transform(X2)

    return [(X1, y1, "Moons"), (X2, y2, "Blobs with varying density")]

# Visualize the results
def visualize_dbscan(datasets, eps_values, min_samples_values):
    fig, axes = plt.subplots(len(datasets), len(eps_values) + 1, figsize=(15, 8))

    for i, (X, y_true, title) in enumerate(datasets):
        # Plot original data
        axes[i, 0].scatter(X[:, 0], X[:, 1], c=y_true, cmap='viridis', s=20)
        axes[i, 0].set_title(f"{title}\n(True Labels)")
        axes[i, 0].set_xticks([])
        axes[i, 0].set_yticks([])

        # Plot DBSCAN results with different parameters
        for j, (eps, min_samples) in enumerate(zip(eps_values, min_samples_values), 1):
            dbscan = DBSCAN(eps=eps, min_samples=min_samples)
            labels = dbscan.fit_predict(X)

            axes[i, j].scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=20)
            axes[i, j].set_title(f"eps={eps}, min_samples={min_samples}")
            axes[i, j].set_xticks([])
            axes[i, j].set_yticks([])

    plt.tight_layout()
    plt.show()

# Helper function to help choose eps parameter
def find_optimal_eps(X, k=5):
    """Plot k-distance graph to help choose optimal eps parameter"""
    neigh = NearestNeighbors(n_neighbors=k)
    nbrs = neigh.fit(X)
    distances, indices = nbrs.kneighbors(X)
    distances = np.sort(distances[:, k-1], axis=0)

    plt.figure(figsize=(8, 5))
    plt.plot(distances)
    plt.xlabel('Points sorted by distance')
    plt.ylabel(f'{k}-NN distance')
    plt.title('K-Distance Graph for EPS estimation')
    plt.grid(True)
    plt.show()

# Example usage
if __name__ == "__main__":
    # Generate sample datasets
    datasets = generate_datasets()

    # Try different parameter combinations
    eps_values = [0.3, 0.3, 0.3]
    min_samples_values = [5, 10, 20]

    # Visualize the results
    visualize_dbscan(datasets, eps_values, min_samples_values)

    # Show how to find optimal eps for the first dataset
    print("Finding optimal EPS for moons dataset:")
    find_optimal_eps(datasets[0][0])
```

```python
# Install required library (uncomment if needed)
# !pip install hdbscan

import hdbscan
import numpy as np
from sklearn.datasets import make_moons
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Generate sample data
X, _ = make_moons(n_samples=1000, noise=0.032, random_state=42)
X = StandardScaler().fit_transform(X)

# Apply HDBSCAN clustering
clusterer = hdbscan.HDBSCAN(
    min_cluster_size=200,      # Smallest grouping to consider a cluster
    min_samples=3,           # Controls conservativeness of clustering
    gen_min_span_tree=True   # Enable minimum spanning tree generation
)
cluster_labels = clusterer.fit_predict(X)

# Visualize results
plt.scatter(X[:, 0], X[:, 1], c=cluster_labels, cmap='viridis', s=10)
plt.title('HDBSCAN Clustering Results')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.colorbar(label='Cluster Label')
plt.show()

# Print cluster information
print(f"Number of clusters found: {clusterer.labels_.max() + 1}")
print(f"Points classified as noise: {np.sum(clusterer.labels_ == -1)}")
```
